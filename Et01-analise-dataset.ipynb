```python
import pandas as pd

# Criando um DataFrame de exemplo
data = {'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Idade': [25, 30, 35, 28, 22],
        'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'São Paulo', 'Rio de Janeiro']}

df = pd.DataFrame(data)

print("DataFrame original:")
display(df)

# Exemplos de busca de dados:

# 1. Selecionando uma coluna específica
print("\nSelecionando a coluna 'Nome':")
display(df['Nome'])

# 2. Selecionando múltiplas colunas
print("\nSelecionando as colunas 'Nome' e 'Cidade':")
display(df[['Nome', 'Cidade']])

# 3. Filtrando linhas com base em uma condição (idade maior que 28)
print("\nFiltrando linhas com idade maior que 28:")
display(df[df['Idade'] > 28])

# 4. Filtrando linhas com base em múltiplas condições (cidade 'São Paulo' e idade menor que 30)
print("\nFiltrando linhas com cidade 'São Paulo' e idade menor que 30:")
display(df[(df['Cidade'] == 'São Paulo') & (df['Idade'] < 30)])

# 5. Buscando por valor específico em uma coluna (quem mora no Rio de Janeiro)
print("\nBuscando quem mora no Rio de Janeiro:")
display(df[df['Cidade'] == 'Rio de Janeiro'])

# 6. Usando .loc para selecionar por rótulo (selecionando a primeira linha pelo índice)
print("\nSelecionando a primeira linha usando .loc:")
display(df.loc[0])

# 7. Usando .iloc para selecionar por posição (selecionando a segunda linha pela posição)
print("\nSelecionando a segunda linha usando .iloc:")
display(df.iloc[1])
```
